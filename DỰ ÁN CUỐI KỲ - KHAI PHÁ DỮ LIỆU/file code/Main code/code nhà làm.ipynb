We can make a quick function to help impute values based on the weights of the currently known values

````python
# write a function that takes a variable and returns a randomly selected 
# value if the variable is 'unknown'
def impute_values(variable, values, prob):
    if variable == 'unknown':
        return np.random.choice(values, p=prob)
    else: 
        return variable 
````

## **IV.a.Giaỉ thích - làm rõ**

**Ở đây, ta sẽ áp dụng các phương pháp xử lý dữ liệu nâng cao để cải thiện hiệu suất của mô hình.**
- ở chương rồi, ta đã nêu ra các vấn đề cần khắc phục, ta sẽ áp dụng các biện pháp khắc phục trong chương này. Xử lý xong ta sẽ kiểm tra lại rồi mới xác nhận tới bước lựa chọn và xây dựng mô hình

**câu hỏi:** Vì Sao Phải Chia Train / Validation / Test Trước Khi SMOTE, Feature Selection, PCA, Chuẩn hóa, One-Hot?

Trước khi áp dụng các kỹ thuật tiền xử lý nâng cao (SMOTE, Feature Selection, PCA, biến đổi log, chuẩn hóa, One-Hot Encoding), cần **chia dữ liệu thành Train / Validation / Test** để tránh **data leakage** và đánh giá mô hình một cách trung thực.

1. Tránh Data Leakage (Rò rỉ dữ liệu)

Nếu ta dùng **toàn bộ dữ liệu** để:
- Tính scaler (chuẩn hóa)
- Tính PCA
- Chọn feature (feature selection)
- Chạy SMOTE

rồi **sau đó mới** tách Train/Test, thì các thống kê của **Test set đã bị “nhìn thấy” trước** (thông qua mean, std, ma trận tương quan, vector PCA, v.v.).

Hậu quả:
- Mô hình học được thông tin từ Test set một cách gián tiếp.
- Kết quả trên Test/Validation quá đẹp nhưng **không phản ánh đúng thực tế**.

2. Nguyên tắc đúng: Fit trên Train, Apply lên Validation/Test

Quy trình chuẩn:

1. **Chia dữ liệu thành:**
   - Train
   - Validation
   - Test

2. **Chỉ dùng Train để:**
   - Fit StandardScaler / MinMaxScaler  
   - Fit One-Hot Encoder  
   - Fit PCA  
   - Chọn feature (feature selection)  
   - Áp dụng SMOTE (cân bằng lớp)

3. **Dùng các đối tượng đã fit từ Train để transform:**
   - Validation
   - Test

Như vậy:
- Validation/Test được xử lý **bằng cùng phép biến đổi**, nhưng **không góp phần vào việc học các phép biến đổi đó**.
- Đảm bảo chúng là “dữ liệu chưa từng thấy” (unseen data) đúng nghĩa.

4. Liên hệ với từng kỹ thuật

- **SMOTE:**  
  Nếu áp dụng trước khi chia, các mẫu tổng hợp sẽ “rò rỉ” sang Test → mô hình đánh giá quá lạc quan.

- **Feature Selection & PCA:**  
  Nếu chọn feature/PCA trên full data, kết quả lựa chọn đã sử dụng thông tin từ Test. Đúng ra, việc chọn feature phải dựa trên Train (hoặc Train+Validation) rồi **giữ cố định** cho Test.

- **Biến đổi Logarit & Chuẩn hóa:**  
  Mean, std, min, max… phải được ước lượng trên Train. Test chỉ được transform bằng các giá trị đó, không được “góp phần” tính toán.

- **One-Hot Encoding:**  
  Tập các category nên được học từ Train. Validation/Test chỉ map theo các category đã biết; category lạ được xử lý riêng (nếu có).


**Lý do ta chia train/test/validation**
1. Khác nhau giữa train–test và train–validation–test:

    Chia train–test (2 tập):

        Train: dùng để học mô hình / khám phá quan hệ.

        Test: dùng để đánh giá cuối cùng.

        Không có tập riêng để:

        chọn mô hình tốt nhất,

        Chủ yếu minh họa, không tối ưu nhiều.
        Chia train–validation–test (3 tập)
        Train: huấn luyện mô hình, fit scaler, encoder, SMOTE, PCA...

    Validation:
        chọn mô hình (model selection),

        chỉ dùng 1 lần cuối để đánh giá khách quan,

````python
import pandas as pd
from sklearn.model_selection import train_test_split

# Giả định: 'data' là DataFrame đã được tải và tiền xử lý cơ bản (đã có cột 'y_encoded')

# =================================================================
# 1. PHÂN CHIA DỮ LIỆU THÀNH 3 TẬP: TRAIN, VALIDATION, TEST
# =================================================================

# Tách biến mục tiêu (Target) và biến độc lập (Features)
X = data.drop('y_encoded', axis=1)
y = data['y_encoded']

# Bước 1: Chia Test Set (ví dụ 15% tổng dữ liệu) và giữ lại tập tạm (Train + Validation)
# Sử dụng stratify=y để đảm bảo tỷ lệ lớp mục tiêu được giữ nguyên
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42, stratify=y
)

# Bước 2: Chia Train và Validation từ X_temp (85% dữ liệu còn lại)
# Để Validation chiếm 17% tổng dữ liệu, ta cần tính tỷ lệ từ X_temp: 17% / 85% = 0.2
test_size_for_validation = 0.20 # 20% của tập X_temp (0.20 * 0.85 = 0.17)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=test_size_for_validation, random_state=42, stratify=y_temp
)

# =================================================================
# 2. KIỂM TRA KẾT QUẢ
# =================================================================

total_samples = X.shape[0]

print("✅ Chia dữ liệu hoàn tất theo chiến lược Train/Validation/Test")
print("-" * 50)
print(f"Tổng số mẫu: {total_samples}")
print(f"1. Tập Train (Huấn luyện): {X_train.shape[0]} mẫu ({round(X_train.shape[0]/total_samples*100)}%)")
print(f"2. Tập Validation (Tinh chỉnh): {X_val.shape[0]} mẫu ({round(X_val.shape[0]/total_samples*100)}%)")
print(f"3. Tập Test (Đánh giá cuối): {X_test.shape[0]} mẫu ({round(X_test.shape[0]/total_samples*100)}%)")

# Kiểm tra tỷ lệ lớp mục tiêu (để xác nhận stratify hoạt động)
print("\nSo sánh tỷ lệ lớp thiểu số ('y_encoded' = 1):")
print(f"  Train: {y_train.mean():.4f}")
print(f"  Validation: {y_val.mean():.4f}")
print(f"  Test: {y_test.mean():.4f}")
````

````python
# 1. Biến phân loại (One-Hot Encoding)
categorical_features = [
    'job', 'marital', 'education', 'default', 'housing',
    'loan', 'contact', 'month', 'day_of_week', 'poutcome'
]

# 2. Biến cần Biến đổi Logarit (thường là biến > 0 và lệch)
log_features = ['age', 'campaign', 'previous'] 

# 3. Biến số còn lại (Chỉ cần Chuẩn hóa)
numerical_features = [
    'emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 
    'euribor3m', 'nr.employed', 'p_contacted'
]
````

## **IV.c.Mã Hóa Biến Phân Loại & Biến Đổi Logarit**

````python
# =================================================================
# BƯỚC 1: ONE-HOT ENCODING (Biến phân loại)
# BƯỚC 2 & 3: BIẾN ĐỔI LOGARIT & CHUẨN HÓA (Biến số)
# =================================================================

# Transformer cho biến Logarit: Log(1+x) và Chuẩn hóa
log_transformer = Pipeline(steps=[
    # np.log1p là Log(1+x), được dùng để xử lý các giá trị 0 an toàn.
    ('log', FunctionTransformer(np.log1p, validate=True)), 
    ('scaler', StandardScaler()) # Chuẩn hóa Z-Score
])

# Transformer cho biến số còn lại: Chỉ Chuẩn hóa
standard_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# Transformer cho biến phân loại: One-Hot Encoding
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Kết hợp các bước xử lý ban đầu
preprocessor = ColumnTransformer(
    transformers=[
        ('log_scale', log_transformer, log_features),
        ('standard_scale', standard_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='drop' 
)

# Áp dụng FIT & TRANSFORM trên tập Train
X_train_processed = preprocessor.fit_transform(X_train)

print("✅ Bước 1, 2, 3 hoàn tất: One-Hot, Logarit, Chuẩn hóa (X_train_processed)")
print(f"Kích thước X_train sau tiền xử lý: {X_train_processed.shape}")
````

## **IV.d.Xử Lý Đa Cộng Tuyến (PCA)**

````python
# =================================================================
# BƯỚC 4: PCA (Xử lý Đa cộng tuyến)
# =================================================================

# Khởi tạo PCA: Giữ lại 95% phương sai của dữ liệu
pca_model = PCA(n_components=0.95, random_state=42)

# FIT (học) PCA trên X_train đã được tiền xử lý
pca_model.fit(X_train_processed)

# TRANSFORM (áp dụng) PCA
X_train_pca = pca_model.transform(X_train_processed)

print("-" * 60)
print("✅ Bước 4 hoàn tất: PCA")
print(f"Số lượng đặc trưng ban đầu (sau One-Hot): {X_train_processed.shape[1]}")
print(f"Số lượng thành phần chính (PC) được giữ lại: {X_train_pca.shape[1]}")
print(f"Tổng phương sai được giữ lại: {pca_model.explained_variance_ratio_.sum():.4f}")
````

## **IV.e.Xử Lý Mất Cân Bằng Lớp (SMOTE)**

````python
# =================================================================
# BƯỚC 5: SMOTE (Xử lý Mất cân bằng lớp)
# =================================================================

# Khởi tạo SMOTE
smote_sampler = SMOTE(sampling_strategy='minority', random_state=42)

# Áp dụng SMOTE trên tập Train đã qua PCA
X_train_final, y_train_final = smote_sampler.fit_resample(
    X_train_pca, y_train
)

print("-" * 60)
print("✅ Bước 5 hoàn tất: SMOTE")
print(f"Kích thước tập Train cuối cùng (X_train_final): {X_train_final.shape}")
print(f"Số lượng mẫu lớp '0' sau SMOTE: {len(y_train_final) - y_train_final.sum()}")
print(f"Số lượng mẫu lớp '1' sau SMOTE: {y_train_final.sum()}")

print("\nBộ dữ liệu đã sẵn sàng để huấn luyện mô hình!")
````

## **IV.f.Tại sao không tách One-Hot Encoding thành một phần riêng?**

 **Lý Do Tích Hợp One-Hot Encoding (OHE) Trong ColumnTransformer Cùng Với Các Biến Đổi Khác**

Việc gom OHE, biến đổi Logarit và Chuẩn hóa vào một `ColumnTransformer` duy nhất thay vì tách riêng lẻ được khuyến nghị vì ba lý do chính:

---

1. Tối ưu hóa Quy trình và Tính nhất quán (Pipeline)

  - **Tính nhất quán:**  
    `ColumnTransformer` được fit trên tập huấn luyện (`X_train`) một lần duy nhất. Khi áp dụng lên tập Validation/Test (`X_val`, `X_test`), chỉ cần gọi `.transform()` để đảm bảo các tham số (ví dụ: danh sách danh mục OHE, mean/std của chuẩn hóa) hoàn toàn giống nhau, tránh sai lệch.

  - **Tiết kiệm code:**  
    Giảm số dòng code cho các bước tiền xử lý riêng lẻ, giúp pipeline tổng thể gọn gàng, dễ bảo trì và tái sử dụng.

---

2. Khả năng Xử lý Dữ liệu Hỗn Hợp (ColumnTransformer)

  - **Tách biệt nhiệm vụ xử lý:**  
    Cho phép áp dụng các phương pháp khác nhau cho từng nhóm cột:  
    - OHE chỉ áp dụng cho biến phân loại (categorical features).  
    - Biến đổi Logarit và Chuẩn hóa chỉ áp dụng cho biến số (numerical features).

  - **Giảm phức tạp:**  
    Tránh phải xử lý thủ công việc nối kết đầu ra của OHE (thường là ma trận thưa) với các biến số đã được xử lý khác, giúp quy trình rõ ràng và hiệu quả hơn.

---

3. Quan trọng nhất: Tránh Rò rỉ Dữ liệu (Data Leakage) 🛡️

  - **Rò rỉ trong Mã hóa:**  
    Nếu tách OHE ra và fit trên toàn bộ dữ liệu (bao gồm cả Validation/Test), mô hình sẽ biết trước tất cả các danh mục từ tập Validation/Test, gây rò rỉ thông tin.  
    Khi OHE nằm trong `ColumnTransformer` và pipeline, nó chỉ được fit trên tập Train, giữ cho Validation/Test hoàn toàn "mù" về thông tin này.

  - **Toàn vẹn trong Cross-Validation:**  
    Khi dùng GridSearchCV hoặc cross_val_score, pipeline tự động fit tiền xử lý (OHE, Logarit, Chuẩn hóa, PCA...) chỉ trên tập Train của từng fold, rồi transform tập Validation fold đó.  
    Nếu OHE bị tách riêng, bạn phải tự quản lý thủ công việc áp dụng đúng cho từng fold, dễ dẫn đến lỗi và rò rỉ dữ liệu.

---

**Tóm Lại**

  `ColumnTransformer` giúp tự động hóa, đồng bộ hóa các bước tiền xử lý khác nhau một cách an toàn và hiệu quả, đảm bảo:

  - Tính nhất quán trong biến đổi dữ liệu giữa các tập Train/Validation/Test.  
  - Tránh rò rỉ dữ liệu nguy hiểm làm sai lệch kết quả mô hình.  
  - Giúp quy trình pipeline gọn gàng, dễ bảo trì và mở rộng.

## **IV.g.Kiểm tra vấn đề giải quyết xong chưa**

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.outliers_influence import variance_inflation_factor
from scipy.sparse import issparse # Thêm thư viện để kiểm tra loại sparse matrix

# =================================================================
# 1. KIỂM TRA MẤT CÂN BẰNG LỚP (SMOTE) ⚖️
# =================================================================
# Kiểm tra tỷ lệ lớp trong tập huấn luyện cuối cùng

print("="*60)
print("KIỂM TRA 1: MẤT CÂN BẰNG LỚP (SMOTE)")

# Tính toán số lượng mẫu của từng lớp
so_mau_lop_0 = len(y_train_final) - y_train_final.sum()
so_mau_lop_1 = y_train_final.sum()
ty_le_can_bang = so_mau_lop_1 / so_mau_lop_0

print(f"Số lượng mẫu Lớp '0' sau SMOTE: {so_mau_lop_0}")
print(f"Số lượng mẫu Lớp '1' sau SMOTE: {so_mau_lop_1}")
print(f"Tỷ lệ Lớp 1 / Lớp 0: {ty_le_can_bang:.4f}")

if 0.99 <= ty_le_can_bang <= 1.01:
    print("-> ✅ Đã giải quyết: Dữ liệu đã được cân bằng gần như 1:1.")
else:
    print("-> ❌ Lỗi: SMOTE chưa đạt được tỷ lệ 1:1. Cần kiểm tra lại tham số sampling_strategy.")
# =================================================================
# HÀM HỖ TRỢ XỬ LÝ DỮ LIỆU SPARSE/DENSE
# =================================================================
def ensure_dense_array(X):
    """Kiểm tra và chuyển đổi dữ liệu thành NumPy array dày nếu cần."""
    if issparse(X):
        return X.toarray()
    return X

# =================================================================
# 2. KIỂM TRA ĐA CỘNG TUYẾN (PCA / VIF CHECK) 📉
# =================================================================
# PCA đã được tính toán trên X_train_pca. Ta chỉ cần kiểm tra
print("\n" + "="*60)
print("KIỂM TRA 2: ĐA CỘNG TUYẾN (PCA)")

# Sử dụng hàm hỗ trợ để đảm bảo X_train_pca là array dày
X_vif_check_array = ensure_dense_array(X_train_pca)
X_vif_check = pd.DataFrame(X_vif_check_array) 

# ... (Phần code tính VIF giữ nguyên)
if X_vif_check.shape[0] > 10000:
    X_vif_sample = X_vif_check.sample(10000, random_state=42)
else:
    X_vif_sample = X_vif_check.copy()

vif_data = pd.DataFrame()
vif_data["feature"] = X_vif_sample.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif_sample.values, i) for i in range(X_vif_sample.shape[1])]

vif_max = vif_data['VIF'].max()
so_pc_vif_cao = len(vif_data[vif_data['VIF'] > 10])

print(f"VIF Tối đa của các Thành phần chính (PC): {vif_max:.2f}")
print(f"Số lượng PC có VIF > 10: {so_pc_vif_cao}")

if so_pc_vif_cao == 0 and vif_max < 5:
    print("-> ✅ Đã giải quyết: PCA đã làm giảm Đa cộng tuyến hiệu quả (VIF thấp).")
else:
    print("-> ⚠️ Cảnh báo: Vẫn còn PC có VIF cao. Cần xem xét giảm n_components.")


# =================================================================
# 3. KIỂM TRA PHÂN BỐ LỆCH & CHUẨN HOÁ (LOGARIT & SCALING) 📉
# =================================================================
print("\n" + "="*60)
print("KIỂM TRA 3: PHÂN BỐ LỆCH & CHUẨN HOÁ (Logarit & Scaling)")

# Sử dụng hàm hỗ trợ cho X_train_processed
X_processed_array = ensure_dense_array(X_train_processed)
X_processed_df = pd.DataFrame(X_processed_array)

# ... (Phần code kiểm tra Mean/Std và vẽ biểu đồ giữ nguyên)
col_check = X_processed_df.columns[0] 
mean_check = X_processed_df.iloc[:, col_check].mean()
std_check = X_processed_df.iloc[:, col_check].std()

print(f"Kiểm tra phân bố của cột đầu tiên (Mean & Std):")
print(f"Giá trị Trung bình (Mean): {mean_check:.4f}")
print(f"Độ lệch chuẩn (Std Dev): {std_check:.4f}")

# Trực quan hóa
plt.figure(figsize=(7, 3))
sns.histplot(X_processed_df.iloc[:, col_check], kde=True)
plt.title('Phân bố Cột sau Logarit và Chuẩn hóa')
plt.xlabel('Giá trị chuẩn hóa')
plt.show() 

if abs(mean_check) < 0.01 and abs(std_check - 1) < 0.01:
    print("-> ✅ Đã giải quyết: Dữ liệu đã được chuẩn hóa thành công (Mean ~ 0, Std Dev ~ 1).")
else:
    print("-> ❌ Lỗi: Có thể có lỗi trong quá trình Chuẩn hóa.")

# =================================================================
# 4. KIỂM TRA BIẾN PHÂN LOẠI (ONE-HOT ENCODING) 🔢
# =================================================================
print("\n" + "="*60)
print("KIỂM TRA 4: BIẾN PHÂN LOẠI (ONE-HOT ENCODING)")

so_luong_dac_trung_ohe = X_processed_df.shape[1]
so_luong_dac_trung_pca = X_vif_check_array.shape[1]

print(f"Số lượng đặc trưng sau OHE/Logarit/Chuẩn hóa: {so_luong_dac_trung_ohe}")
print(f"Số lượng đặc trưng sau PCA: {so_luong_dac_trung_pca}")

# Giả định X_train là DataFrame gốc
if 'X_train' in globals() and X_vif_check_array.shape[1] > X_train.shape[1]: 
    print(f"-> ✅ Đã giải quyết: Số lượng cột đã tăng lên, xác nhận OHE đã hoạt động.")
    print("-> Kết luận: Tất cả dữ liệu hiện tại đều là số và đã được chuẩn hóa.")
else:
    print("-> ❌ Lỗi: OHE có thể chưa được áp dụng đúng hoặc X_train chưa được định nghĩa.")
````

````python
# =================================================================
# ÁP DỤNG CÁC BƯỚC TIỀN XỬ LÝ CHO TẬP VALIDATION
# =================================================================

# 1. Áp dụng Preprocessor (Logarit, Chuẩn hóa, One-Hot) lên X_val
# Lưu ý: Chỉ dùng .transform(), không dùng .fit_transform()
X_val_processed = preprocessor.transform(X_val)

# 2. Áp dụng PCA (Sử dụng pca_model đã fit trên X_train_processed)
# Lưu ý: Chỉ dùng .transform()
X_val_final = pca_model.transform(X_val_processed)

print("✅ Hoàn tất chuyển đổi tập Validation.")
print(f"Kích thước X_val sau xử lý (X_val_final): {X_val_final.shape}")
````

Kiểm Tra Xây Dựng Mô Hình (Logistic Regression)
Bây giờ, chúng ta sẽ huấn luyện một mô hình cơ bản (ví dụ: Logistic Regression) trên dữ liệu đã qua SMOTE và đánh giá trên tập Validation đã qua xử lý.

````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score, f1_score, accuracy_score, confusion_matrix
import pandas as pd # Cần thiết để hiển thị Confusion Matrix đẹp hơn

# Giả định X_train_final, y_train_final, X_val_final, y_val là các biến đã được định nghĩa
# =================================================================
# XÂY DỰNG MÔ HÌNH VÀ KIỂM TRA
# =================================================================

# 1. Khởi tạo mô hình
model_check = LogisticRegression(random_state=42, solver='liblinear')

# 2. Huấn luyện mô hình (trên dữ liệu đã qua SMOTE)
model_check.fit(X_train_final, y_train_final)

# 3. Dự đoán trên tập Validation
y_val_pred = model_check.predict(X_val_final)
y_val_proba = model_check.predict_proba(X_val_final)[:, 1] # Xác suất cho lớp dương (1)

# =================================================================
# 4. ĐÁNH GIÁ KẾT QUẢ KIỂM TRA
# =================================================================
print("\n" + "="*50)
print("ĐÁNH GIÁ MÔ HÌNH KIỂM TRA TRÊN TẬP VALIDATION")
print("Dữ liệu đã qua: Logarit, Chuẩn hóa, One-Hot, PCA, SMOTE")
print("="*50)

# 1. Thêm Accuracy Score
print(f"Accuracy Score: {accuracy_score(y_val, y_val_pred):.4f}")

# 2. Thêm Confusion Matrix
print("\nConfusion Matrix:")
cm = confusion_matrix(y_val, y_val_pred)
# Hiển thị Confusion Matrix rõ ràng hơn
cm_df = pd.DataFrame(cm, 
                     index = ['Actual Negative (0)', 'Actual Positive (1)'], 
                     columns = ['Predicted Negative (0)', 'Predicted Positive (1)'])
print(cm_df)

# 3. Các chỉ số quan trọng khác
print(f"\nROC-AUC Score: {roc_auc_score(y_val, y_val_proba):.4f}")
print(f"F1 Score: {f1_score(y_val, y_val_pred):.4f}")
print("\nClassification Report (bao gồm Precision/Recall/F1):\n")
print(classification_report(y_val, y_val_pred))

# Kết luận kiểm tra
if roc_auc_score(y_val, y_val_proba) > 0.65:
    print("\n🎉 Dữ liệu đã được xử lý thành công! Mô hình đã có khả năng dự đoán.")
else:
    print("\n⚠️ Mô hình chưa hiệu quả, cần xem xét tinh chỉnh tham số (Hyperparameter Tuning) hoặc thử các mô hình phức tạp hơn.")
````

| KNN                | Đơn giản, không cần huấn luyện                          | Chậm khi dự đoán trên dữ liệu lớn, nhạy cảm với tỷ lệ thuộc tính |
| XGBoost            | Hiệu suất vượt trội, xử lý phi tuyến tốt, điều chỉnh mạnh mẽ | Nhạy cảm với tham số, thời gian huấn luyện lâu              |

 B. Gom Cụm (Clustering - Học không Giám sát)

| Thuật Toán           | Ưu Điểm                                              | Nhược Điểm                                                   |
|----------------------|------------------------------------------------------|--------------------------------------------------------------|

3. Thang Đo Đánh Giá (Metrics)

 A. Thang Đo Phân Lớp (Classification Metrics)

| Thang Đo      | Công Thức / Khái Niệm                                   | Tác Động & Chiến Lược                                        |
|---------------|---------------------------------------------------------|--------------------------------------------------------------|
| Accuracy      | \[ \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN} \]                | Không nên làm mục tiêu chính trên dữ liệu mất cân bằng        |
| Confusion Matrix | Ma trận TP, FP, TN, FN                                  | Giúp trực quan hóa FP để chọn mô hình giảm thiểu lãng phí    |
| Precision (Lớp 1) | \[ \text{Precision} = \frac{TP}{TP + FP} \]                              | Tiêu chí scoring trong GridSearchCV để giảm FP, tăng hiệu quả chi phí Marketing |
| Recall (Lớp 1) | \[ \text{Recall} = \frac{TP}{TP + FN} \]                                | Theo dõi để tránh giảm quá nhiều khi tăng Precision          |
| F1 Score      | \[ F1 = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}} \] | Tiêu chí scoring để cân bằng giữa chi phí (Precision) và khả năng tìm kiếm (Recall) |

 B. Thang Đo Gom Cụm (Clustering Metrics)

| Thang Đo          | Công Thức / Khái Niệm                                   | Tác Động & Chiến Lược                                        |
|-------------------|---------------------------------------------------------|--------------------------------------------------------------|
| Silhouette Score  | Đo độ tương đồng điểm với cụm của nó so với cụm khác \([-1,1]\) | Chọn số cụm tối ưu, đảm bảo phân đoạn rõ ràng, có ý nghĩa kinh doanh |
| Inertia           | Tổng bình phương khoảng cách điểm tới tâm cụm gần nhất  | Dùng Elbow method để xác định k tại điểm "khuỷu tay"         |

````python
from sklearn.cluster import MiniBatchKMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
import warnings
import numpy as np
import pandas as pd
import time

# Tắt cảnh báo
warnings.filterwarnings('ignore') 

# -----------------------------------------------------------------------------
# CẤU HÌNH TỐI ƯU HIỆU NĂNG (QUAN TRỌNG)
# -----------------------------------------------------------------------------
# Giả định X_train_final đã có
X_cluster = X_train_final

# GIỚI HẠN MẪU ĐỂ TĂNG TỐC ĐỘ
# Silhouette Score rất nặng (O(N^2)), chỉ nên tính trên khoảng 10k mẫu max
SILHOUETTE_SAMPLE_SIZE = 10000 

# Agglomerative Clustering tốn cực nhiều RAM, chỉ nên train trên khoảng 3k-5k mẫu
AGG_MAX_SAMPLES = 5000

k_range = range(2, 7) 

# Tạo mẫu dữ liệu rút gọn để tính toán nhanh
if len(X_cluster) > SILHOUETTE_SAMPLE_SIZE:
    X_sil_sample = shuffle(X_cluster, random_state=42, n_samples=SILHOUETTE_SAMPLE_SIZE)
else:
    X_sil_sample = X_cluster

if len(X_cluster) > AGG_MAX_SAMPLES:
    X_agg_sample = shuffle(X_cluster, random_state=42, n_samples=AGG_MAX_SAMPLES)
else:
    X_agg_sample = X_cluster

# Lưu trữ kết quả
inertia_scores = []
kmeans_silhouette_scores = []
agg_silhouette_scores = []

print("="*75)
print("BẮT ĐẦU PHÂN TÍCH CLUSTERING (PHIÊN BẢN 'TURBO')")
print(f"- Dữ liệu gốc: {X_cluster.shape}")
print(f"- Mẫu tính Silhouette: {X_sil_sample.shape}")
print(f"- Mẫu chạy Hierarchical: {X_agg_sample.shape}")
print("="*75)

# =============================================================================
# 1. CLUSTERING LOOP
# =============================================================================

for k in k_range:
    start_time = time.time()
    
    # --- A. MiniBatch KMeans (Nhanh hơn KMeans thường rất nhiều) ---
    # batch_size càng lớn càng chính xác nhưng tốn RAM hơn
    kmeans = MiniBatchKMeans(n_clusters=k, random_state=42, batch_size=2048, n_init=3)
    kmeans.fit(X_cluster) # Có thể fit trên toàn bộ dữ liệu vì MiniBatch rất nhanh
    
    inertia_scores.append(kmeans.inertia_)
    
    # Tính Silhouette Score (Chỉ tính trên mẫu rút gọn!)
    if k > 1:
        # Dự đoán nhãn cho tập mẫu để tính score
        labels_sample = kmeans.predict(X_sil_sample)
        k_score = silhouette_score(X_sil_sample, labels_sample)
        kmeans_silhouette_scores.append(k_score)
    
    # --- B. Hierarchical Clustering (Agglomerative) ---
    # Chỉ chạy trên X_agg_sample vì chạy full sẽ tràn RAM/rất lâu
    agg = AgglomerativeClustering(n_clusters=k)
    agg_labels = agg.fit_predict(X_agg_sample)
    
    if k > 1:
        # Tính Score trực tiếp trên mẫu đã train
        a_score = silhouette_score(X_agg_sample, agg_labels)
        agg_silhouette_scores.append(a_score)

    print(f"✅ Hoàn thành K={k} trong {time.time() - start_time:.2f}s")

# =============================================================================
# 2. TỔNG HỢP KẾT QUẢ
# =============================================================================

results_cluster = pd.DataFrame({
    'K': list(k_range),
    'KMeans_Inertia': inertia_scores,
    'KMeans_Silhouette': kmeans_silhouette_scores,
    'Agg_Silhouette_Sample': agg_silhouette_scores # Lưu ý đây là score trên sample
})

# Tìm K tối ưu
best_idx_kmeans = results_cluster['KMeans_Silhouette'].idxmax()
best_kmeans_k = results_cluster.loc[best_idx_kmeans, 'K']
max_kmeans_sil = results_cluster.loc[best_idx_kmeans, 'KMeans_Silhouette']

best_idx_agg = results_cluster['Agg_Silhouette_Sample'].idxmax()
best_agg_k = results_cluster.loc[best_idx_agg, 'K']
max_agg_sil = results_cluster.loc[best_idx_agg, 'Agg_Silhouette_Sample']

print("\n" + "="*75)
print("KẾT QUẢ ĐÁNH GIÁ (Dựa trên mẫu rút gọn)")
print("="*75)
print(results_cluster.to_string(index=False))

print("\n" + f"🏆 K-Means tốt nhất: K = {int(best_kmeans_k)} (Score: {max_kmeans_sil:.4f})")
print(f"🏆 Hierarchical tốt nhất: K = {int(best_agg_k)} (Score: {max_agg_sil:.4f})")

# Chọn mô hình cuối cùng
if max_kmeans_sil >= max_agg_sil:
    print(f"\n=> CHỌN MODEL: MiniBatchKMeans (K={best_kmeans_k})")
    # Train lại model tốt nhất trên toàn bộ dữ liệu
    final_model = MiniBatchKMeans(n_clusters=int(best_kmeans_k), random_state=42, batch_size=4096).fit(X_cluster)
else:
    print(f"\n=> CHỌN MODEL: Hierarchical (K={best_agg_k})")
    print("Lưu ý: Hierarchical khó áp dụng trên data lớn để predict, cân nhắc dùng KMeans thay thế nếu cần tốc độ.")
    # Hierarchical không có method predict cho data mới, nên vẫn dùng dữ liệu mẫu
    final_model = AgglomerativeClustering(n_clusters=int(best_agg_k)).fit(X_agg_sample)
````
